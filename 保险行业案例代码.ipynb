### 探索空值的个数和比例是多少

### 分析这些空值的特点，确定填充策略

#### 空值填充策略：
分类变量：通常分类水平出现的次数多，出现频率就是最高的，用众数填充。
数值型变量：幸福指数、收入所处排名是分类变量

### 根据我们的策略编写函数进行空值填充

```python
#去掉年龄，其他的用众数填充
NA[NA.Var!='age'].Var
```

```python
df[NA[NA.Var!='age'].Var[0]].mode()[0]
```

```python
#对列名进行遍历，依次进行填充
for i in NA[NA.Var!='age'].Var:
       df[i].fillna(df[i].mode()[0],inplace=True)
```

```python
#年龄采用均值填充
df.age.fillna(df.age.mean(),inplace=True)
```

```python
df.isnull().sum()    #查看是否有那列还有空值
```

## 数据建模


```python
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:,1:]
y = df['resp_flag']
```

### 切分数据集


```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,y,test_size=0.3,
                                          random_state=420)
```

### 查看模型基础效果


```python
clf=tree.DecisionTreeClassifier().fit(Xtrain,Ytrain)
clf.score(Xtest,Ytest)
```

### 模型调参

```python
#尝试使用最小叶节点样本数量和最小分割样本数量进行调参
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'min_samples_leaf':list(range(1000,6000,100)),
             'min_samples_split':list(range(4000,6000,100))}
GR = GridSearchCV(tree.DecisionTreeClassifier(),param_grid,cv=5)
GR.fit(Xtrain,Ytrain)

GR.best_params_  #运行最好结果leaf 1500 ; split 4000
```

```python
#使用最优参数，重新建模
clf = tree.DecisionTreeClassifier(criterion='gini',max_depth=4,
                                 min_samples_leaf=1500,
                                 min_samples_split=4000)
clf.fit(Xtrain,Ytrain)
clf.score(Xtest,Ytest)
```

```python
import graphviz
```

```python
dot_data=tree.export_graphviz(clf,feature_names=features,
                             class_names=['NP','P'],
                                 filled=True,   #填充颜色
                             rounded=True)   #弧度
```
